# groupby

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
```

```python
drinks.beer_servings.mean()
```

#### Agrupamos

```python
drinks.groupby('continent').beer_servings.mean()
```

```python
drinks[drinks.continent=='Africa'].beer_servings.mean()
```

### Funciones de agregaci√≥n. mean(), max(), min(), agg(['count','min'])

```python
drinks.groupby('continent').beer_servings.agg(['mean', 'max', 'count'])
```

```python
drinks.groupby('continent').mean()
```

```python
%matplotlib inline
```

```python
drinks.groupby('continent').mean().plot(kind='bar')
```

# Trabajar con valores desaparecidos (NaN)

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.tail()
```

```python
ufo.isnull().tail()
# Produce un True para los que cumplen la condici√≥n, esto es para los que son nulos (NaN)
```

```python
ufo.notnull().tail()
```

```python
ufo.isnull().sum()
# obtiene el n√∫mero de nulos por cada columna
```

```python
ufo[ufo.City.isnull()].shape
```

##  Opciones: qu√© hacer con valores desaparecidos

### 1. Borrar registros con NaN

```python
ufo.dropna().shape

# elimina cualquier fila que tenga al menos un NaN en sus resgistros, es temporal, sin inplace=True no guarda los cambios
```

```python
ufo.dropna(how='all').shape

# borra los registros en los cuales TODOS los valores sean NaN
```

```python
# Borrar subconjuntos de valores

ufo.dropna(subset=['City', 'Shape Reported']).shape
```

```python
ufo['Shape Reported'].value_counts()

# cuenta la frecuencia por cada figura, pero s√≥lo si no es NaN
```

```python
ufo['Shape Reported'].value_counts(dropna=False)
#incluye aquellos donde los valores est√°n desaparecidos (NaN)
```

#### Cambiar los valores de NaN por otro


```python
ufo['Shape Reported'].fillna(value='VARIOUS', inplace=True)
```

```python
ufo['Shape Reported'].value_counts(dropna=False)
```
