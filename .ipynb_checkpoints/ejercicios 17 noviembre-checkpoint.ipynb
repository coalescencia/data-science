# Ejercicio 4.6

```python
def longest_of(list_of_words):
    length_so_far = 0

    for word in list_of_words:
        this_length = len(word)
        if this_length > length_so_far:
            length_so_far = this_length
            
    return length_so_far

print(longest_of(['man', 'woman']))
print(longest_of([]))
print(longest_of(['man', 'woman', 'mimuyhg']))
```

```python
list_of_words = ['man', 'woman', 'mimuyhg']
print(list(map(len,list_of_words)))
#list comprehension
[len(word) for word in list_of_words]
```

```python
#otra forma de hacer la funcion
def longest_of(list_of_words):
    lengths = [len(word) for word in list_of_words]
    if len(list_of_words) == 0:
        longest = None
    else:
        longest = max(lengths)
        
    
             
    return longest

print(longest_of(['man', 'woman']))
print(longest_of([]))
print(longest_of(['man', 'woman', 'mimuyhg']))
```

### Ejercicio 4.3

```python
list_of_words = ['tocoto','man', 'woman', 'mimuyhg']
def names_with(list_of_names, character):
    names = []
    
    for name in list_of_names:
        if character in name:
            names.append(name)
            
    return names

print(names_with(list_of_words, 'o'))
print(names_with([], 'o'))
```

```python
#ejercicio 4.1 de las hojas de teoria python
list_of_words = ['tocoto','man', 'woman', 'mimuyhg', 'tail']
resultados_deseados = {'m' : ['man', 'mimuyhg'], 't' : ['tail','tocoto'], 'w' : ['woman']}
```

```python
def categorize(list_of_words):
    dictionary = dict()
    
    for word in list_of_words:
        first = word[0]
        
        if first not in dictionary:
            dictionary[first] = [word]
        else:
            dictionary[first].append(word)
        
    return dictionary
        

categorize(list_of_words)
```

```python
#ejercicio 4.2 hojas teoria python
# set devuelve los valores únicos, elimina duplicados
list_of_words = ['tocoto','man', 'woman', 'money', 'tail']
def sort_by_variety(list_of_words):
    return sorted(list_of_words, key = lambda w: len(set(w)))


sort_by_variety(list_of_words)
```
