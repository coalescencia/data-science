# Filtrar filas de un data frame por valores de columna

### Filtros 1ª forma:

```python
peliculas.shape
```

```python
boleanos = []
for longitud in peliculas.duration:
    if longitud >= 200:
        boleanos.append(True)
    else:
        boleanos.append(False)
```

```python
boleanos[0:5]
```

```python
len(boleanos)
```

```python
is_long = pd.Series(boleanos)
```

```python
is_long.head()
```

```python
peliculas[is_long]
```

### Filtros 2ª forma:

```python
peliculas
```

```python
is_long = peliculas.duration >= 200
```

```python
is_long.head()
```

```python
peliculas[is_long]
```

```python
# forma abreviada:

peliculas[peliculas.duration >=200]
```

```python
peliculas[peliculas.duration >=200]['genre']
```

```python
peliculas[peliculas.duration >=200].genre[peliculas.genre =='Drama']
```

```python
peliculas.loc[peliculas.duration >=200, 'genre']
```

## Aplicar múltiples criterios de filtro a un Data frame

```python
peliculas = pd.read_csv('http://bit.ly/imdbratings')
```

```python
peliculas.head()
```

```python
peliculas[peliculas.duration >=200]
```

```python
# peliculas[peliculas.duration >=200].genre[peliculas.genre =='Drama']

# más de un filtro:

peliculas[(peliculas.duration >=200) & (peliculas.genre =='Drama')]
```

```python

peliculas[(peliculas.duration >=200) | (peliculas.genre =='Drama')]
```

```python
# filtrar por varios valores de una misma columna:

peliculas[peliculas.genre.isin(['Crime', 'Drama', 'Action'])]
```

```python

peliculas[(peliculas.genre.isin(['Crime', 'Drama', 'Action']))&(peliculas.duration >=180)]
```

```python
for c in peliculas.genre.unique(): #retorna valores no repetidos de genre
    print(c)
    
```

```python
for index, row in peliculas.iterrows():
    print(index, row.genre, row.duration)
```
